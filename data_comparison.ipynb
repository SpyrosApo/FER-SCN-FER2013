```python
import os
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot  as plt
```

```python
with open('test_labels.txt', 'r') as f:
    test_labels = f.read().splitlines()
#Put the output txt of the model in the line of code below
with open('test_0.txt', 'r') as f:
    result_labels = f.read().splitlines()
print ('True labels:')
print (len(test_labels))
print ('Precicted labels:')
print (len( result_labels))
correct_counter=0
counter=0
for f, b in zip(test_labels,  result_labels):
    if f==b:
          correct_counter+=1
    counter+=1
print('Number of correct labels:')
print(correct_counter)
```

```python
accuraccy=correct_counter/len(test_labels)
print('Number of wrong labels:')
print(len(test_labels)-correct_counter)
print("accuraccy:",accuraccy,"%" )
```

```python
labels=['Angry','Disgust','Fear','Happy','Sad','Surprise','Neutral']
conf_matrix = confusion_matrix(y_true=test_labels, y_pred= result_labels)
#
# Print the confusion matrix using Matplotlib
#
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(conf_matrix, cmap=plt.cm.Blues)
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax.text(x=j, y=i,s=conf_matrix[i, j], va='center', ha='center', size='xx-large')
 
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels)
plt.show()
```
